```python
# Normalize via l2 norm
def normalize(x, cols):
    for col in cols:
        col_sum = x[col].sum()
        for index, row in x.iterrows():
            st = row[col]
            new = st/col_sum
            x.loc[index, [col]] = new
    return x
updated_df = normalize(updated_df, ["author.playtime_at_review", "author.playtime_forever", "author.playtime_last_two_weeks"])

updated_df
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

text = "I love you so much"

s = TextBlob("I love you so much")
print("Textblob:", s.sentiment)

s1 = SentimentIntensityAnalyzer()
sentiment_dict = s1.polarity_scores(text)
print("VADER:", sentiment_dict)
if sentiment_dict['compound'] >= 0.05 :
    print("Positive")

elif sentiment_dict['compound'] <= - 0.05 :
    print("Negative")

else :
    print("Neutral")
```

```python
updated_df = df[:100000]
updated_df
```

```python
polarityT = [TextBlob(str(review)).polarity for review in updated_df["review"]]
updated_df["polarityT"] = polarityT
updated_df
```

```python
polarityV = []
polarityV_label = []
s = SentimentIntensityAnalyzer()
for review in updated_df["review"]:
    sent_dict = s.polarity_scores(str(review))
    polarityV.append(sent_dict['compound'])
    
    if sent_dict['compound'] >= 0.05:
        polarityV_label.append("Positive")
    elif sent_dict['compound'] <= - 0.05:
        polarityV_label.append("Negative")
    else:
        polarityV_label.append("Neutral")

updated_df["polarityV"] = polarityV
updated_df["polarityV_label"] = polarityV_label
updated_df
```

```python
#data = updated_df[["votes_helpful", "votes_funny", "polarity"]]
data = updated_df[["polarityV"]]
labels = updated_df["recommended"]

X = np.array(data)
Y = np.array(labels)
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.25)

model = KNeighborsClassifier(n_neighbors=11)

model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
print(predicted)
```
