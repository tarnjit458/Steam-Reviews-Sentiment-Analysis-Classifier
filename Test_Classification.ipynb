```python
#! pip install autocorrect
#! pip install pyspellchecker
from autocorrect import Speller
from spellchecker import SpellChecker
import time
import re

# Note: Takes a long time
# Not sure if its worth it
start = time.time()
updated_df = df[:100]
# updated_df

def autocorrect(x):
    # Perform spell checking and auto correct misspelled words
    check = Speller(lang='en', fast=True)
    for index, row in x.iterrows():
        n = check(row['review'])
        x.loc[index, ['review']] = str(n)
    return x

def textblob(x):
    for index, row in x.iterrows():
        blob = TextBlob(row['review'])
        n = blob.correct()
        # n = check(row['review'])
        x.loc[index, ['review']] = str(n)
    return x

spell=Speller(lang="en")
WORD = re.compile(r'\w+')
def reTokenize(doc):
    tokens = WORD.findall(doc)
    return tokens
def spell_correct(text):
    sptext = []
    for doc in text:
        sptext.append(' '.join([spell(w).lower() for w in reTokenize(doc)]))      
    return sptext  

def pyspell(x):
    for index, row in x.iterrows():
        st = row['review']
        st = reTokenize(st)
        n = spell_correct(st)
        # n = check(row['review'])
        x.loc[index, ['review']] = str(n)
    return x

#pyspell.correction('gud')

updated_df = pyspell(updated_df)
end = time.time()
print(end - start)
updated_df
```

```python
polarity = [TextBlob(str(review)).polarity for review in updated_df["review"]]
updated_df["polarity"] = polarity
updated_df
```

```python
data = updated_df[["votes_helpful", "votes_funny", "polarity"]]
labels = updated_df["recommended"]

X = np.array(data)
Y = np.array(labels)
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.25)

model = KNeighborsClassifier(n_neighbors=3)

model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
print(predicted)
```
